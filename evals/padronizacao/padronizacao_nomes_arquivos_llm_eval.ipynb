```python
import pandas as pd
import shutil
from pathlib import Path

lims_path = Path("../avaliacao_llms")

items = []
for item in sorted(lims_path.iterdir()):
    if item.is_dir():
        nome_atual = item.name
        novo_nome = nome_atual
        novo_nome = novo_nome.replace("__anthropic__", "__")
        novo_nome = novo_nome.replace("__openai__", "__")
        novo_nome = novo_nome.replace("__deepseek__deepseek-", "__deepseek-")
        novo_nome = novo_nome.replace("__", "_")

        items.append({"nome_atual": nome_atual, "novo_nome": novo_nome})

df = pd.DataFrame(items)
df
```

```python
for _, row in df.iterrows():
    caminho_atual = lims_path / row["nome_atual"]
    caminho_novo = lims_path / row["novo_nome"]

    if caminho_atual == caminho_novo:
        print(f"➖ Já correto: {row['nome_atual']}")
    elif caminho_atual.exists() and not caminho_novo.exists():
        shutil.move(str(caminho_atual), str(caminho_novo))
        print(f"✓ Renomeado: {row['nome_atual']} → {row['novo_nome']}")
    elif not caminho_atual.exists() and caminho_novo.exists():
        print(f"✓ Já renomeado: {row['novo_nome']}")
    else:
        print(f"⚠️ Conflito: {row['nome_atual']} e {row['novo_nome']} ambos existem!")
```

```python
items_files = []

for _, row in df.iterrows():
    pasta = lims_path / row["novo_nome"]

    for arquivo in sorted(pasta.iterdir()):
        if arquivo.suffix == ".xlsx":
            nome_atual = arquivo.name
            avaliador = nome_atual.split("_")[-2]
            novo_nome_arquivo = f"{row['novo_nome']}_{avaliador}_eval.xlsx"

            items_files.append({
                "pasta": row["novo_nome"],
                "nome_atual": nome_atual,
                "novo_nome": novo_nome_arquivo
            })

df_files = pd.DataFrame(items_files)
df_files
```

```python
for _, row in df_files.iterrows():
    caminho_atual = lims_path / row["pasta"] / row["nome_atual"]
    caminho_novo = lims_path / row["pasta"] / row["novo_nome"]

    if caminho_atual == caminho_novo:
        print(f"➖ Já correto: {row['nome_atual']}")
    elif caminho_atual.exists() and not caminho_novo.exists():
        shutil.move(str(caminho_atual), str(caminho_novo))
        print(f"✓ Renomeado: {row['nome_atual']} → {row['novo_nome']}")
    elif not caminho_atual.exists() and caminho_novo.exists():
        print(f"✓ Já renomeado: {row['novo_nome']}")
    elif caminho_atual.exists() and caminho_novo.exists():
        print(f"⚠️ Conflito: ambos existem!")
    else:
        print(f"✗ Não encontrado: {row['nome_atual']}")
```

```python
grupos = {}

for _, row in df_files.iterrows():
    arquivo = lims_path / row["pasta"] / row["novo_nome"]
    nome = row["novo_nome"]
    
    if nome.endswith("opus_eval.xlsx"):
        avaliador = "opus"
    elif nome.endswith("gpt_eval.xlsx"):
        avaliador = "gpt"
    else:
        continue
    
    if arquivo.exists():
        df_temp = pd.read_excel(arquivo, nrows=0)
        colunas = tuple(df_temp.columns.tolist())
        
        if avaliador not in grupos:
            grupos[avaliador] = []
        grupos[avaliador].append({"arquivo": nome, "colunas": colunas})

for grupo, items in grupos.items():
    colunas_unicas = set(i["colunas"] for i in items)
    print(f"\n{'='*50}")
    print(f"Grupo: {grupo} ({len(items)} arquivos)")
    if len(colunas_unicas) == 1:
        print(f"✓ Todos compartilham as mesmas colunas:")
        for col in list(colunas_unicas)[0]:
            print(f"   - {col}")
    else:
        print(f"⚠️ Colunas divergentes!")
        for item in items:
            print(f"\n  {item['arquivo']}:\n    {list(item['colunas'])}")
```
