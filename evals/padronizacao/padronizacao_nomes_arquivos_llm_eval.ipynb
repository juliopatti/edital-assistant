```python
import pandas as pd
import shutil
from pathlib import Path

lims_path = Path("../avaliacao_llms")

items = []
for item in sorted(lims_path.iterdir()):
    if item.is_dir():
        nome_atual = item.name
        novo_nome = nome_atual
        novo_nome = novo_nome.replace("__anthropic__", "__")
        novo_nome = novo_nome.replace("__openai__", "__")
        novo_nome = novo_nome.replace("__deepseek__deepseek-", "__deepseek-")
        novo_nome = novo_nome.replace("__", "_")

        items.append({"nome_atual": nome_atual, "novo_nome": novo_nome})

df = pd.DataFrame(items)
df
```

```python
for _, row in df.iterrows():
    caminho_atual = lims_path / row["nome_atual"]
    caminho_novo = lims_path / row["novo_nome"]

    if caminho_atual == caminho_novo:
        print(f"➖ Já correto: {row['nome_atual']}")
    elif caminho_atual.exists() and not caminho_novo.exists():
        shutil.move(str(caminho_atual), str(caminho_novo))
        print(f"✓ Renomeado: {row['nome_atual']} → {row['novo_nome']}")
    elif not caminho_atual.exists() and caminho_novo.exists():
        print(f"✓ Já renomeado: {row['novo_nome']}")
    else:
        print(f"⚠️ Conflito: {row['nome_atual']} e {row['novo_nome']} ambos existem!")
```

```python
items_files = []

for _, row in df.iterrows():
    pasta = lims_path / row["novo_nome"]

    for arquivo in sorted(pasta.iterdir()):
        if arquivo.suffix == ".xlsx":
            nome_atual = arquivo.name
            avaliador = nome_atual.split("_")[-2]
            novo_nome_arquivo = f"{row['novo_nome']}_{avaliador}_eval.xlsx"

            items_files.append({
                "pasta": row["novo_nome"],
                "nome_atual": nome_atual,
                "novo_nome": novo_nome_arquivo
            })

df_files = pd.DataFrame(items_files)
df_files
```

```python
for _, row in df_files.iterrows():
    caminho_atual = lims_path / row["pasta"] / row["nome_atual"]
    caminho_novo = lims_path / row["pasta"] / row["novo_nome"]

    if caminho_atual == caminho_novo:
        print(f"➖ Já correto: {row['nome_atual']}")
    elif caminho_atual.exists() and not caminho_novo.exists():
        shutil.move(str(caminho_atual), str(caminho_novo))
        print(f"✓ Renomeado: {row['nome_atual']} → {row['novo_nome']}")
    elif not caminho_atual.exists() and caminho_novo.exists():
        print(f"✓ Já renomeado: {row['novo_nome']}")
    elif caminho_atual.exists() and caminho_novo.exists():
        print(f"⚠️ Conflito: ambos existem!")
    else:
        print(f"✗ Não encontrado: {row['nome_atual']}")
```

```python
grupos = {}

for _, row in df_files.iterrows():
    arquivo = lims_path / row["pasta"] / row["novo_nome"]
    nome = row["novo_nome"]
    
    if nome.endswith("opus_eval.xlsx"):
        avaliador = "opus"
    elif nome.endswith("gpt_eval.xlsx"):
        avaliador = "gpt"
    else:
        continue
    
    if arquivo.exists():
        df_temp = pd.read_excel(arquivo, nrows=0)
        colunas = tuple(df_temp.columns.tolist())
        
        if avaliador not in grupos:
            grupos[avaliador] = []
        grupos[avaliador].append({"arquivo": nome, "colunas": colunas})

for grupo, items in grupos.items():
    colunas_unicas = set(i["colunas"] for i in items)
    print(f"\n{'='*50}")
    print(f"Grupo: {grupo} ({len(items)} arquivos)")
    if len(colunas_unicas) == 1:
        print(f"✓ Todos compartilham as mesmas colunas:")
        for col in list(colunas_unicas)[0]:
            print(f"   - {col}")
    else:
        print(f"⚠️ Colunas divergentes!")
        for item in items:
            print(f"\n  {item['arquivo']}:\n    {list(item['colunas'])}")
```

```python
for pasta in df_files["pasta"].unique():
    arquivos = df_files[df_files["pasta"] == pasta]
    
    caminho_gpt = lims_path / pasta / arquivos[arquivos["novo_nome"].str.endswith("gpt_eval.xlsx")]["novo_nome"].values[0]
    caminho_opus = lims_path / pasta / arquivos[arquivos["novo_nome"].str.endswith("opus_eval.xlsx")]["novo_nome"].values[0]
    
    df_gpt = pd.read_excel(caminho_gpt)[["id", "categoria", "pergunta", "resposta", "justificativa_gpt", "avaliacao_gpt"]]
    df_opus = pd.read_excel(caminho_opus)[["justificativa_opus", "avaliacao_opus"]]
    
    df_consolidado = pd.concat([df_gpt, df_opus], axis=1)
    df_consolidado["convergencia"] = df_consolidado["avaliacao_gpt"] == df_consolidado["avaliacao_opus"]
    
    print(f"\n{'='*50}")
    print(f"{pasta}")
    
    vc = df_consolidado["convergencia"].value_counts()
    vc_norm = df_consolidado["convergencia"].value_counts(normalize=True).round(2)
    print(pd.concat([vc, vc_norm], axis=1, keys=["count", "%"]))
    
    soma_gpt = df_consolidado["avaliacao_gpt"].sum()
    soma_opus = df_consolidado["avaliacao_opus"].sum()
    max_score = len(df_consolidado) * df_consolidado["avaliacao_gpt"].max()
    
    print(f"\n  avaliacao_gpt:  soma={soma_gpt} | {soma_gpt/max_score*100:.1f}%")
    print(f"  avaliacao_opus: soma={soma_opus} | {soma_opus/max_score*100:.1f}%")
```

```python
df_files["modelo"] = df_files["pasta"].str.split("_", n=1).str[1]

for modelo in df_files["modelo"].unique():
    arquivos = df_files[df_files["modelo"] == modelo]
    
    dfs = []
    for _, row in arquivos[arquivos["novo_nome"].str.endswith("gpt_eval.xlsx")].iterrows():
        pasta = row["pasta"]
        caminho_gpt = lims_path / pasta / row["novo_nome"]
        caminho_opus = lims_path / pasta / arquivos[(arquivos["pasta"] == pasta) & (arquivos["novo_nome"].str.endswith("opus_eval.xlsx"))]["novo_nome"].values[0]
        
        df_gpt = pd.read_excel(caminho_gpt)[["id", "categoria", "pergunta", "resposta", "justificativa_gpt", "avaliacao_gpt"]]
        df_opus = pd.read_excel(caminho_opus)[["justificativa_opus", "avaliacao_opus"]]
        df_consolidado = pd.concat([df_gpt, df_opus], axis=1)
        df_consolidado["convergencia"] = df_consolidado["avaliacao_gpt"] == df_consolidado["avaliacao_opus"]
        dfs.append(df_consolidado)
    
    df_modelo = pd.concat(dfs, ignore_index=True)
    
    print(f"\n{'='*50}")
    print(f"{modelo} ({len(dfs)} editais, {len(df_modelo)} linhas)")
    
    vc = df_modelo["convergencia"].value_counts()
    vc_norm = df_modelo["convergencia"].value_counts(normalize=True).round(2)
    print(pd.concat([vc, vc_norm], axis=1, keys=["count", "%"]))
    
    soma_gpt = df_modelo["avaliacao_gpt"].sum()
    soma_opus = df_modelo["avaliacao_opus"].sum()
    max_score = len(df_modelo) * df_modelo["avaliacao_gpt"].max()
    
    print(f"\n  avaliacao_gpt:  soma={soma_gpt} | {soma_gpt/max_score*100:.1f}%")
    print(f"  avaliacao_opus: soma={soma_opus} | {soma_opus/max_score*100:.1f}%")
```

```python
rows = []

df_files["modelo"] = df_files["pasta"].str.split("_", n=1).str[1]

for modelo in df_files["modelo"].unique():
    arquivos = df_files[df_files["modelo"] == modelo]
    
    dfs = []
    for _, row in arquivos[arquivos["novo_nome"].str.endswith("gpt_eval.xlsx")].iterrows():
        pasta = row["pasta"]
        caminho_gpt = lims_path / pasta / row["novo_nome"]
        caminho_opus = lims_path / pasta / arquivos[(arquivos["pasta"] == pasta) & (arquivos["novo_nome"].str.endswith("opus_eval.xlsx"))]["novo_nome"].values[0]
        
        df_gpt = pd.read_excel(caminho_gpt)[["avaliacao_gpt"]]
        df_opus = pd.read_excel(caminho_opus)[["avaliacao_opus"]]
        df_consolidado = pd.concat([df_gpt, df_opus], axis=1)
        df_consolidado["convergencia"] = df_consolidado["avaliacao_gpt"] == df_consolidado["avaliacao_opus"]
        dfs.append(df_consolidado)
    
    df_modelo = pd.concat(dfs, ignore_index=True)
    max_score = len(df_modelo) * df_modelo["avaliacao_gpt"].max()
    
    rows.append({
        "modelo": modelo,
        "n_linhas": len(df_modelo),
        "convergencia_count": df_modelo["convergencia"].sum(),
        "convergencia_%": round(df_modelo["convergencia"].mean() * 100, 1),
        "soma_gpt": df_modelo["avaliacao_gpt"].sum(),
        "score_gpt_%": round(df_modelo["avaliacao_gpt"].sum() / max_score * 100, 1),
        "soma_opus": df_modelo["avaliacao_opus"].sum(),
        "score_opus_%": round(df_modelo["avaliacao_opus"].sum() / max_score * 100, 1),
    })

df_modelos = pd.DataFrame(rows)
df_modelos
```
