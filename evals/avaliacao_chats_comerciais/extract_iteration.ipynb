```python
from pathlib import Path
import re
import pandas as pd
from bs4 import BeautifulSoup

HTML_PATH = Path("data/bndes_std_claude.html")

OUT_XLSX = HTML_PATH.with_name("bndes_std_claude_qa.xlsx")
OUT_CSV = HTML_PATH.with_name("bndes_std_claude_qa.csv")

def clean_text(el):
    txt = el.get_text("\n", strip=True)
    txt = re.sub(r"\n{3,}", "\n\n", txt)
    txt = re.sub(r"\n([,.;:!?])", r"\1", txt)
    return txt.strip()

html = HTML_PATH.read_text(encoding="utf-8", errors="ignore")
soup = BeautifulSoup(html, "html.parser")

messages = []

for el in soup.find_all(True):
    if el.get("data-testid") == "user-message":
        messages.append({
            "autor": "user",
            "texto": clean_text(el),
            "html": str(el),
        })

    elif el.get("data-is-streaming") == "false":
        resp = el.select_one(".font-claude-response")
        if resp:
            messages.append({
                "autor": "assistant",
                "texto": clean_text(resp),
                "html": str(resp),
            })

rows = []
i = 0
qid = 1

while i < len(messages):
    if messages[i]["autor"] == "user":
        pergunta = messages[i]
        resposta = None

        if i + 1 < len(messages) and messages[i + 1]["autor"] == "assistant":
            resposta = messages[i + 1]
            i += 2
        else:
            i += 1

        rows.append({
            "id": qid,
            "pergunta": pergunta["texto"],
            "resposta": resposta["texto"] if resposta else "",
            "pergunta_html": pergunta["html"],
            "resposta_html": resposta["html"] if resposta else "",
        })
        qid += 1
    else:
        i += 1

df = pd.DataFrame(rows)

# manter só as colunas desejadas
df = df[["pergunta", "resposta"]]

# ignorar primeira linha
df = df.iloc[1:].reset_index(drop=True)

# recriar id de 1 até N
df.insert(0, "id", range(1, len(df) + 1))

# opcional: garantir que ficou de 1 a 50
print(df["id"].min(), df["id"].max(), len(df))

df.to_excel(OUT_XLSX, index=False)
df.to_csv(OUT_CSV, index=False, encoding="utf-8-sig")

print(f"{len(df)} pares pergunta/resposta extraídos")
print(OUT_XLSX)
print(OUT_CSV)

display(df.head(1))
display(df.tail(1))
```
